# How to Measure Electrical Resistivity
---
You will be using the pcb mounted sample labeled CMP-102 and the Keithley 2450 SourceMeter to measure the electrical resistivity of a sample of silicon. On page 3-2 of the CMP Electrical Transport manual you will find some useful figures. 

We have attached jumper wires to the connector in order to take readings. See the table below

| Port | Color |
| :- | :- |
| A | Yellow |
| B | Brown |
| C | White |
| D | Green |
| P | N/A |
| Q | Red |
| R | Orange |
| S | N/A |

## SourceMeter Setup and Data Taking
You will be using the Keithley SourceMeter (SM) to perform a sweep. It will supply a specific current and read the voltage in response, then repeat those steps for each value for current specified in the settings menu. To set up your data runs do the following steps:

1. Making sure that no connections are made between the SM and the semi-conductor chip, power on the SM
2. Press the "Function" button on the front panel
    - Under the "Sourcing Current and Measure:" heading select "Voltage" Click past the Warning Number 5076
3. Press the "Menu" button on the front panel
    - Under the "Measure" heading select "Settings"
        - Change the setting from "2-Wire" to "4-Wire"
    - Press the "Home" button on the front panel
4. Press the "Menu" button on the front panel
    - Under the "Measure" heading select "Reading Buffers"
        - Create a new reading buffer and name it so you will remember which data run you are on
        - Make sure that the "Make Active" button on the screen is greyed out.
    - Press the "Home" button on the front panel
5. Press the "Menu" button on the front panel
    - Under the "Source" heading select "Sweep"
        - Set the following parameters in this menu
            - Set "Start" to 0 $\mu$A
            - Set "Stop" to 500 $\mu$A
            - Set "Step" to 10 $\mu$A scroll down
            - Set "Source Delay" to 0.1s
            - Set "Source Limit" to 2V
        - Select "Generate"
    - Press the "Home" button on the front panel
6. Connect the leads from the "Force" ports to the A and D jumper wires
7. Connect the leads from the "Sense" ports to the jumper wires corresponding to your desired testing points on the semi-conductor chip
    - In this step you will measure the longitudinal potential difference in order to calculate the resistivity. On your first sweep start by measuring the potential difference between connection points B and C
8. Use the code block below to Trigger the sweep, as the Trigger button on the Source Meter is unresponsive.  

Next you will take multiple individual data runs changing how you connect the sense leads to the semiconductor chip. In order to save the voltage data from multiple sweeps you will need to repeat step 4, make sure you create a new reading buffer for each data run and that your new reading buffer is the active one. After that you will need to regenerate your sweep. So follow step 5, confirm that your settings haven't changed, and select "Generate" to tell the device to save your data in your newly created reading buffer. 

---
## Data Transfer
Once you have taken all of your data for the electrical resistivity portion, you will need to use a USB drive to transfer the data to a computer so you can perform the necessary calculations.

1. Insert the USB drive into the port on the front panel of the SM
2. Press the "Menu" button on the front panel
    - Under the "Measure" heading select "Reading Buffers"
    - Select the name of your current reading buffer
    - In the pop-up menu labeled "Reading Buffer" select the buffer you would like to save
    - Select save to USB
    - In this step you can choose a different name for the file that is saved to your USB drive
    - In the confirmation pop-up screen confirm that your file name is correct and taht it is saving as a .csv file
3. Repeat step 2 for all of the buffers you have data saved in from your measurments of electrical resistivity
4. Press the "Home" button on the front panel
5. It is now safe to remove the USB drive from the SM and transfer the data for analysis

When you are finished taking data for the day make sure to power down the SM and remove any connections made to the semi-conductor chip.

# How to Measure Hall Voltage
---
In order to measure the Hall Voltage of a sample of a semi-conductor you need to apply a magnetic field. Use the red plastic magnetic chamber with two embedded magnets to supply this field. If you powered down the Keithley SourceMeter (SM) after you took your electrical resistivity data you will need to go through the full set up process again. If not, you can skip right to step 4 and create a new reading buffer to save this data into. Then complete step 5 to regenerate your sweep so it will save to the correct reading buffer. In step 7 you will now be using different pairs of connection points on the sample to measure the transverse potential difference on the chip this time. Pair up Q and B, R and C for these measurments making sure to repeat the sweep to take data for both directions across these pairs of connections.

## Measuring the Magnetic Field
For your data analysis you will also need a reading for the field applied to the sample. For this step use the Sypris/F.W. Bell Teslameter. The magnetic probe only measures the field in one direction, so make sure that you spin the probe tip while it is in the magnet chamber and record the maximum value displayed.

1. Power on the Teslameter
    - The meter will calibrate itself
2. Remove the protective cover from the magnet probe
3. Hold the probe so the tip (marked with a white dot) is in the magnet chamber roughly at the center of the magnets above and below
4. Write down the measured value of the field
    - Make sure to take note of the units that the field is being measured in
    
When you are finished taking data for the day make sure to power down the Teslameter, SM, and remove any connections made to the semi-conductor chip.

# Pyvisa
This code uses the package Pyvisa, so that will need to be installed. This first block of code is used to tell what devices are connected to your computer. Then find your device and connect to it directly using the .open_resource() function

```python
import pyvisa

rm = pyvisa.ResourceManager()
print(rm)
connections = rm.list_resources()
print(connections)
```

```python
Keithley = rm.open_resource(connections[0])

print(Keithley.query('*IDN?'))
```

For information on the pyvisa library and the functions called above (ResourceManager(), list_resources(), and open_resource()) [here](https://pyvisa.readthedocs.io/en/latest/) is the link to the documentation. The '\*IDN?' and all other commands either queried or written to the Source Meter see the reference manual which can be downloaded [here](https://www.tek.com/en/support/product-support?series=2400%20Graphical%20Series%20SMU&type=manual)
