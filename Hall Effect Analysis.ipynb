# Linear Fit Hall Voltage

## For Chip 1:

```python
# subtract No B-Field y-axis data from B-Field y-axis data to get Hall voltage
Hall_V_C1_BC = 1E3*(BF_C1_BC_y - ST_C1_BC_y) # for B to C
Hall_V_C1_BQ = 1E3*(BF_C1_BQ_y - ST_C1_BQ_y) # for B to Q
print( "Hall_V_C1_BC: \n", Hall_V_C1_BC.head(), 
        "\nHall_V_C1_BQ: \n", Hall_V_C1_BQ.head())

# do not need to subtract x-axis data because current of B-field / no B-field is the same, rename as Hall current (not necessiary)
Hall_I_C1_BC = BF_C1_BC_x # for B to C
Hall_I_C1_BQ = BF_C1_BQ_x # for B to Q

# Solve the Linear Fit for Chip 1 B to C:
C1_BC_popt, C1_BC_pcov = curve_fit(linFit, Hall_I_C1_BC , Hall_V_C1_BC, p0=[m:=0.75, b:=0]) # solves for m and b
C1_BC_y_fit = linFit(Hall_I_C1_BC, *C1_BC_popt) # creates y(x) = gaus function with fitted values
C1_BC_uncert = np.sqrt(np.diag(C1_BC_pcov))

# Plot Hall Chip 1 B to C:
plt.figure(figsize=(10,5)) # set figure size
plt.xlim(-0.1, 0.6) # x axis limits
plt.ylim(-1, 1) # y axis limits
plt.xlabel("Current, I (mA)") # x axis label
plt.ylabel("Voltage, V (Volts)") # y axis label
plt.title("Chip 1 Hall V vs I: B to C Linear Fit ") # plot title
plt.plot(Hall_I_C1_BC, Hall_V_C1_BC, 'o') #plot data points
plt.plot(Hall_I_C1_BC, C1_BC_y_fit, 'r-') #plot linear fit
plt.show() 
print(" B to C Hall Resistance: ", C1_BC_popt[0], "\u00B1 ", C1_BC_uncert[0]) #print hall resistance & uncertainties

# Solve the Linear Fit for Chip 1 B to Q:
C1_BQ_popt, C1_BQ_pcov = curve_fit(linFit, Hall_I_C1_BQ , Hall_V_C1_BQ, p0=[m:=0.75, b:=0]) # solves for m and b
C1_BQ_y_fit = linFit(Hall_I_C1_BC, *C1_BQ_popt) # creates y(x) = gaus function with fitted values
C1_BQ_uncert = np.sqrt(np.diag(C1_BQ_pcov)) 

# Plot Hall Chip 1 B to Q:
plt.figure(figsize=(10,5)) # set figure size
plt.xlim(-0.1, 0.6) # x axis limits
plt.ylim(-1, 1) # y axis limits
plt.xlabel("Current, I (mA)") # x axis label
plt.ylabel("Voltage, V (Volts)") # y axis label
plt.title("Chip 1 Hall V vs I: B to Q Linear Fit ") # plot title
plt.plot(Hall_I_C1_BQ, Hall_V_C1_BQ, 'o') #plot data points
plt.plot(Hall_I_C1_BQ, C1_BQ_y_fit, 'r-') #plot linear fit                      
plt.show()
print(" B to Q Hall Resistance: ", C1_BQ_popt[0], "\u00B1 ", C1_BQ_uncert[0]) #print hall resistance & uncertainties
```

# Calculate Carrier Density: $ n = \frac{B}{ e  t  R_{H}} $

B = B-Field of magnet = $ 9.01 * 10^{-5} $ Teslas

e = electron charge = $ 1.602176634 * 10^{−19}$ Coulombs 

t = thickness of both chips = 0.5 mm 

$R_{H}$ = Hall Resistance between two points for a chip

## Error Propogation for Carrier Density: 

$\sigma_{n}^2 = (\sigma_{R_{H}} * \frac{\partial n }{\partial R_{H}})^2$ $\longrightarrow$ $\sigma_{n}^2 = (\sigma_{R_{H}} * -\frac{B}{e t R_{H}^2})^2$
   

```python
# define constants
B = 9.01E-5 
e = 1.602176634E-19
t = 0.5

# define Hall resistance for each path of both chips
HR_C1_BC = C1_BC_popt[0] #chip 1, B to C
HR_C1_BQ = C1_BQ_popt[0] #chip 1, B to Q
HR_C2_BC = C2_BC_popt[0] #chip 2, B to C
HR_C2_BQ = C2_BQ_popt[0] #chip 2, B to Q
HR_C2_BR = C2_BR_popt[0] #chip 2, B to R

# define Hall resistance uncertainties for each path of both chips
HRsig_C1_BC = C1_BC_uncert[0] #chip 1, B to C
HRsig_C1_BQ = C1_BQ_uncert[0] #chip 1, B to Q
HRsig_C2_BC = C2_BC_uncert[0] #chip 2, B to C
HRsig_C2_BQ = C2_BQ_uncert[0] #chip 2, B to Q
HRsig_C2_BR = C2_BR_uncert[0] #chip 2, B to R

# define Hall resistance values & Hall resistance uncertainty values as array
HR_values = [ HR_C1_BC, HR_C1_BQ, HR_C2_BC, HR_C2_BQ, HR_C2_BR ]
HR_sig_values = [ HRsig_C1_BC, HRsig_C1_BQ, HRsig_C2_BC, HRsig_C2_BQ, HRsig_C2_BR ]

# define carrier density function
n_values = [B / (e * t * HR) for HR in HR_values]

# define carrier density labels
labels = [ "chip 1 from B to C:", "chip 1 from B to Q:", 
          "chip 2 from B to C:", "chip 2 from B to Q:", "chip 2 from B to R:" ]

#define carrier density uncertainty function, remove neg sign from sqrt or else complex numver
n_sig_values = [np.sqrt(HR_sig * ( B / (e * t * HR**2))) for HR_sig, HR in zip(HR_sig_values, HR_values)]

#print carrier densities and uncertainties
print("Carrier Densities in Ohms per mm^3:\n")
    
for n, n_sig, label in zip( n_values, n_sig_values, labels):
    print(label, n,"\u00B1 ", n_sig)

print("\nNegative carrier densities means the conduction is by holes.",
      "\nPositive carrier densities means the conduction is by electrons.")
```
